```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
os.chdir('C:/data')
```

```python
#append()向列表的尾部添加一个新的元素
#'+='表示为两个值相加，然后返回值给符号左侧的变量
#age(-1到10)，-1代表prefer not to say
#gender(-1到3)，-1代表prefer not to say
#edu(-1到5)，-1代表prefer not to say
ages = []
noage = 0
gends = []
nogend = 0
edus = []
noedu = 0
with open('demo.txt') as f:
    for line in f:
        age,gend,edu = line.split('\t')
        age = int(age)
        gend = int(gend)
        edu = int(edu)
        
        if age>=0:
            ages.append(age)
        else:
            noage += 1
        if gend>=0:
            gends.append(gend)
        else:
            nogend += 1
        if edu>=0:
            edus.append(edu)
        else:
            noedu += 1
```

```python
# fig,ax = plt.subplots(),using the variable ax for single a Axes
#如果是fig,axs = plt.subplots(2,2)则会出现4个图
fig,ax = plt.subplots()
labels = ['prefer not', '<7', '7-12', '12-17', '18-25', '26-35', '36-45', '46-55', '56-65', '66-75', '>76']
#displot在未来的版本中会被删除，可以选用另外的function
sns.distplot(ages, ax=ax, bins=len(labels)-1)
#调整x轴刻度位置，没有这行代码x轴无法显示prefer not和>76
ax.set_xticks(np.arange(0, len(labels))+.5)
#设置x轴标签，且旋转90度，没有这行代码x轴为-1到10
ax.set_xticklabels(labels, rotation=90)
print(noage)
```

```python
fig,ax = plt.subplots()
labels = ['prefer not', 'female', 'male', 'other']
#加上kde=False会把折线去掉
sns.distplot(gends, ax=ax, bins=len(labels)-1, kde=False)
ax.set_xticks(np.arange(0, len(labels))+.5)
ax.set_xticklabels(labels, rotation=90)
print(nogend)
```

```python
fig,ax = plt.subplots()
labels = ['prefer not', 'primary', 'secondary', 'A-level', 'Bachelors', 'Masters+']
sns.distplot(edus, ax=ax, bins=len(labels)-1, kde=False)
ax.set_xticks(np.arange(0, len(labels))+.5)
ax.set_xticklabels(labels, rotation=90)
print(noedu)
```
